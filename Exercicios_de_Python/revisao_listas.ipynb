```python
# Revisão do tópico de Listas em Python!

# Principais métodos de listas:
# 1. append(): Adiciona um elemento ao final da lista.
# 2. len(): Retorna o número de elementos na lista.
# 3. insert(): Insere um elemento em uma posição específica da lista.
# 4. remove(): Remove a primeira ocorrência de um elemento da lista.
# 5. pop(): Remove e retorna o elemento em uma posição específica da lista.
# 6. clear(): Remove todos os elementos da lista.
# 7. index(): Retorna o índice da primeira ocorrência de um elemento na lista.
# 8. count(): Retorna o número de vezes que um elemento aparece na lista.
# 9. sort(): Ordena os elementos da lista.
# 10. reverse(): Inverte a ordem dos elementos na lista.
# 11. extend(): Adiciona os elementos de outra lista ao final da lista atual.
# 12. __add__(): Retorna uma nova lista que é a concatenação da lista atual com outra lista.
# 13. enumerate(): Retorna um objeto enumerado, que é uma sequência de tuplas contendo o índice e o valor de cada elemento na lista.
```

```python
# Oque é uma lista?
# Uma lista em Python é uma coleção ordenada e mutável de elementos.
# As listas são definidas usando colchetes [] e os elementos são 
# separados por vírgulas.
# "Ordenada" significa que os elementos mantêm a posição que você 
# colocou. "Mutável" significa que você pode alterar, adicionar ou 
# remover elementos depois de criada.

# Lembrando que o índice de uma lista começa em 0, ou seja, o primeiro
# elemento está na posição 0, o segundo na posição 1, e assim por diante.
```

```python
# Exemplo de uso dos métodos de listas:

frutas = ['maçã', 'banana', 'laranja', 'abacaxi', 'abacaxi']  # Criando uma lista de frutas
```

```python
frutas.append('uva')  # Adiciona 'uva' ao final da lista
print(frutas)  # Output: ['maçã', 'banana', 'laranja
```

```python
frutas.__len__()  # Retorna o número de elementos na lista (0, pois a lista está vazia) 
```

```python
frutas.insert(1, 'abacaxi')  # Insere 'abacaxi' na posição 1
# As listas em Python são indexadas a partir de 0, então 'abacaxi' 
# será inserido entre 'maçã' e 'banana'.
print(frutas)  # Output: ['maçã[0]', 'abacaxi[1]', 'banana[2]', 'laranja[3]', 'uva[4]']
```

```python
frutas.remove('banana')  # Remove a primeira ocorrência de 'banana' da lista
print(frutas)  # Output: ['maçã', 'abacaxi', 'laranja', 'uva']
```

```python
frutas.pop(2)  # Remove e retorna o elemento na posição 2 (que é 'laranja')
print(frutas)  # Output: ['maçã', 'abacaxi', 'uva']
```

```python
frutas.index('abacaxi')  # Retorna o índice da primeira ocorrência de 'abacaxi' (1)
```

```python
frutas.count('abacaxi')  # Retorna o número de vezes que 'abacaxi' aparece na lista (2) 
```

```python
frutas.sort()  # Ordena os elementos da lista
print(frutas)  # Output: ['abacaxi', 'maçã', 'uva']
```

```python
frutas.reverse()  # Inverte a ordem dos elementos na lista
print(frutas)  # Output: ['uva', 'maçã', 'abacaxi']
```

```python
frutas.extend(['melancia', 'manga'])  # Adiciona os elementos de outra lista ao final da lista atual
print(frutas)  # Output: ['uva', 'maçã', 'abacaxi', 'melancia', 'manga']
```

```python
frutas.__add__(['pera', 'kiwi'])  # Retorna uma nova lista que é a concatenação da lista atual com outra lista
print(frutas)  # Output: ['uva', 'maçã', 'abacaxi', 'melancia', 'manga'] (a lista original permanece inalterada)
print(frutas.__add__(['pera', 'kiwi']))  # Output: ['uva', 'maçã', 'abacaxi', 'melancia', 'manga', 'pera', 'kiwi'] (nova lista resultante da concatenação)
```

```python
fruits = ['grape', 'orange', 'apple']
enumerate_fruits = enumerate(fruits)  # Retorna um objeto enumerado]
print(list(enumerate_fruits))  # Output: [(0, 'grape'), (1, 'orange'), (2, 'apple')] (lista de tuplas com índice e valor)
```

```python
print(type(frutas))  # Output: <class 'list'>
```

```python
print(len(frutas))  # Output: 5 (número de elementos na lista)
print('abacaxi' in frutas)  # Output: True (verifica se 'abacaxi' está na lista)
print(frutas.index('maçã'))  # Output: 1 (índice de 'maçã' na lista)
```

```python
# Oque é Slicing?
# Slicing é uma das funcionalidades mais poderosas das listas. A sintaxe é lista[início:fim:passo] - o início é incluído, o fim é excluído.
print(frutas[1:4])  # Output: ['maçã', 'abacaxi', 'abacaxi'] (sublista do índice 1 ao 3)
```
