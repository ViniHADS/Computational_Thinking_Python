## Métodos de Tuplas

```python
# Método count() para contar quantas vezes um valor aparece na tupla
respostas_forms = ('sim', 'não', 'sim', 'sim', 'não')

print(respostas_forms.count('sim')) # Conta quantas vezes 'sim' aparece na tupla
print(respostas_forms.count('não')) # Conta quantas vezes 'não' aparece na tupla
```

```python
# Método index() para encontrar a posição de um valor na tupla
podio = ('Ana', 'Beatriz', 'Carlos', 'Lucas')

posicao = podio.index('Beatriz') # Encontra a posição de 'Carlos' na tupla

print(f'Beatriz está na posição {posicao} do pódio.') # Imprime a posição de 'Beatriz' no pódio
```

## Funções

```python
# Função que soma os elementos de uma tupla
notas = (8, 9, 7, 10)

total_notas = sum(notas) # Soma os elementos da tupla
print(f'A soma das notas é: {total_notas}') # Imprime a soma das notas
```

```python
# Função max() para encontrar o maior valor em uma tupla
# Função min() para encontrar o menor valor em uma tupla
temperaturas = (15, 28, 22, 19, 30)

print(f'A temperatura mais alta é: {max(temperaturas)}°C') # Imprime a temperatura mais alta da tupla
print(f'A temperatura mais baixa é: {min(temperaturas)}°C') # Imprime a temperatura mais baixa da tupla
```

```python
# Função len() para obter o número de elementos em uma tupla
convidados_festa = ('Alice', 'Bob', 'Charlie', 'Diana')
print(f'O número de convidados na festa é: {len(convidados_festa)}') # Imprime o número de convidados na festa
```

### Exercício: Software de Edição

```python
# Você está criando um app simples para verificar um código de cores RGB, mas os usuários estão inserindo valores fora do intervalo permitido (0-255). Escreva um código que verifica se os valores de cor são válidos e imprime uma mensagem de erro para os valores inválidos.
# Valores de cor RGB devem estar entre 0 e 255. 
# O código abaixo verifica se os valores de cor são válidos e imprime uma mensagem de erro para os valores inválidos.
# Exemplo de valores de cor RGB: (255, 0, 0) para vermelho, (0, 255, 0) para verde, (0, 0, 255) para azul.
cor_errada = (300, 50, -10) # Valores de cor inválidos (RGB fora do intervalo 0-255)

for valor in cor_errada:
    if valor < 0 or valor > 255:
        print(f'Valor de cor inválido: {valor}. Deve estar entre 0 e 255.') # Imprime mensagem de erro para valores de cor inválidos
```

### Exercício: Processamento de Rota de Entrega

```python
# Você está criando um app de processamento de coordenadas de entregas:
# Nessa lista de tupla, lat(é a latitude) e long(é a longitude) estão fora do intervalo permitido para coordenadas geográficas. Escreva um código que verifica se as coordenadas são válidas e imprime uma mensagem de erro para as coordenadas inválidas.
# Coordenadas geográficas válidas: Latitude deve estar entre -90 e 90, Longitude deve estar entre -180 e 180.
# Você também irá contar quantas entregas existem na lista de tuplas usando a função len().
# Verifica se uma coordenada em especifico existe na lista (CD)
# Imprime pro motorista cada coordenada de forma legivel (Latitude: x, Longitude: y)

rota_entregas = [
    (-23.5505, -46.6333),
    (-23.5589, -46.6611),
    (-23.5505, -46.6333), # Entrega duplicada no mesmo prédio
    (-23.5921, -46.6723)
]
```

```python
print(len(rota_entregas)) # Imprime o número de entregas na lista de tuplas
```

```python
for entrega in rota_entregas:
    lat, long = entrega # Desempacotando a tupla em latitude e longitude
    if lat < -90 or lat > 90 or long < -180 or long > 180:
        print(f'Coordenada inválida: Latitude {lat}, Longitude {long}. Deve estar entre -90 e 90 para latitude e -180 e 180 para longitude.') # Imprime mensagem de erro para coordenadas inválidas
    else:
        print(f'Entrega válida: Latitude {lat}, Longitude {long}.') # Imprime mensagem de entrega válida
```

```python
ponto_especifico = (-23.5505, -46.6333)
```

```python
print(rota_entregas.count(ponto_especifico)) # Conta quantas vezes o ponto específico aparece na lista de entregas
```

```python
for destino in rota_entregas:
    print(f'Destino: {destino}') # Imprime cada destino de entrega de forma legível
```

```python
for lat, long in rota_entregas:
    print(f'Latitude: {lat}, Longitude: {long}') # Imprime cada coordenada de entrega de forma legível
```

```python
rota_entregas.append((-23.6123, -46.7456)) # Adiciona uma nova entrega à lista de tuplas (não recomendado, pois tuplas são imutáveis)
```
