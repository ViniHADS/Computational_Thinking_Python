## Tratamento de Dados

#### Normalização de dados cadastrais de CEP

```python
lista_ceps_entrada = [
    '10200-300',
    ' 40500600 ',
    '70800-901',
    '90201.301',
    'Não-Sei'
]

lista_ceps_normalizados = []

for registro_cep in lista_ceps_entrada:
    cep_formatado = registro_cep.strip().replace('-', '').replace('.', '' ).strip()
    lista_ceps_normalizados.append(cep_formatado)
    print(f'Processado: {registro_cep} -> {cep_formatado}')

print(lista_ceps_normalizados)
```

```python
lista_ceps_entrada = [
    '10200-300',
    ' 40500600 ',
    '70800-901',
    '90201.301',
    'Não-Sei',
    '15222.4588',
    '1530054',
    '153468-289'
]

lista_ceps_validados = []
lista_ceps_rejeitados = []

for registro in lista_ceps_entrada:
    # limpeza de dados
    cep_limpo = registro.strip().replace('-', '').replace('.', '' )

    # validação de regra
    # retorna TRUE se na string conter somente digitos
    if len(cep_limpo) == 8 and cep_limpo.isdigit() == True:
        lista_ceps_validados.append(cep_limpo)

    else:
        lista_ceps_rejeitados.append(registro)

print(f'Sucesso: {lista_ceps_validados}')
print(f'Rejeitados: {lista_ceps_rejeitados}')
```
