```python
nomes = ["Alice", "Bob", "Charlie", "David", "Eve"]
print(nomes[0])  # Acessa o primeiro elemento
print(nomes[2])  # Acessa o terceiro elemento
print(nomes[-1]) # Acessa o último elemento
print(nomes[1:4]) # Acessa do segundo ao quarto elemento (exclusivo)
print(nomes[:3])  # Acessa do início ao terceiro elemento (exclusivo)
print(nomes[2:])  # Acessa do terceiro elemento até o final
print(nomes[::2]) # Acessa todos os elementos com passo 2 (pula um elemento)
print(nomes[::-1]) # Acessa todos os elementos em ordem reversa
print(nomes[1:4:2]) # Acessa do segundo ao quarto elemento com passo 2 (pula um elemento)
# print(nomes[5])  # Tenta acessar um índice fora do alcance (gera um erro) #
```

```python
print(frutas[-1])  # Acessa o último elemento
```

```python
lista_linguagens = ["Python", "Java", "C++", "JavaScript", "Ruby"]
print(lista_linguagens[2])  # Acessa o terceiro elemento
```

```python
# Como acessar o primeiro e o último elemento de uma lista
lista_linguagens = ["Python", "Java", "C++", "JavaScript", "Ruby"]
print(lista_linguagens[0])  # Acessa o primeiro elemento
print(lista_linguagens[-1]) # Acessa o último elemento
```

```python
# Acessando apenas 2 elementos específicos
print(lista_linguagens[1])  # Acessa o segundo elemento 
print(lista_linguagens[3])  # Acessa o quarto elemento

# tem como acessar o terceiro elemento e o último elemento ao mesmo tempo?
print(lista_linguagens[2], lista_linguagens[-1])  # Acessa o terceiro e o último elemento

# existe outro metódo para acessar o terceiro e o último elemento ao mesmo tempo?
print(lista_linguagens[2:4:1])  # Acessa do terceiro ao quarto elemento com passo 1 (inclui o terceiro e o quarto elemento)

# diferença entre usar o metódo slice e chamar os elementos diretamente
print(lista_linguagens[2], lista_linguagens[-1])  # Acessa o terceiro e o último elemento diretamente
print(lista_linguagens[2:4:1])  # Acessa do terceiro ao quarto elemento com passo 1 (inclui o terceiro e o quarto elemento)
```

# Funções e Métodos em Listas

### Funções

```python
valores = [10, 5, 20, 15]
print(valores[1])  # Acessa o segundo elemento
```

```python
print(len(valores))  # Imprime o número de elementos na lista
```

```python
print(sum(valores))  # Imprime a soma de todos os elementos da lista
print(max(valores))  # Imprime o valor máximo da lista
print(min(valores))  # Imprime o valor mínimo da lista
```

```python
valores_ordenados = sorted(valores)  # Ordena a lista em ordem crescente
print(valores_ordenados)  # Imprime a lista ordenada
```

```python
valores_ordenados_descrescente = sorted(valores, reverse=True)  # Ordena a lista em ordem decrescente
print(valores_ordenados_descrescente)  # Imprime a lista ordenada em ordem decrescente
```

### Métodos de Listas

```python
# .append() - Adiciona um elemento ao final da lista
valores.append(25)
print(valores)  # Imprime a lista com o novo elemento adicionado

filmes.append("O Senhor dos Anéis")
print(filmes)  # Imprime a lista de filmes com o novo filme adicionado  
```

```python
# .insert() - Adiciona um elemento em uma posição específica da lista
filmes.insert(1, "O Poderoso Chefão")  # Insere o filme "O Poderoso Chefão" na posição 1
print(filmes)  # Imprime a lista de filmes com o novo filme adicionado na posição 1
```

### Remoção de Elementos

```python
# .remove() - Remove a primeira ocorrência de um elemento específico da lista
filmes.remove("O Jogo da Imitação")  # Remove o filme "O Jogo da Imitação" da lista
print(filmes)  # Imprime a lista de filmes com o filme removido
```

```python
# .pop() - Remove um elemento em uma posição específica da lista e retorna o elemento removido
filme_removido = filmes.pop(2)  # Remove o filme na posição 2 e armazena o filme removido
print(filme_removido)  # Imprime o filme que foi removido
```

### Organização e Ordenação

```python
# .sort() - Ordena a lista em ordem crescente (modifica a lista original)
valores.sort()
print(valores)  # Imprime a lista ordenada em ordem crescente

movies = ["Matrix", "O Jogo da Imitação", "Interstellar", "O Senhor dos Anéis", "O Poderoso Chefão"]
movies.sort()
print(movies)  # Imprime a lista de filmes ordenada em ordem alfabética

series = ["Breaking Bad", "Game of Thrones", "Stranger Things", "The Crown", "The Mandalorian"]
series.sort()
print(series)  # Imprime a lista de séries ordenada em ordem alfabética
```

### Busca e Contagem

```python
# .index() - Retorna o índice da primeira ocorrência de um elemento específico na lista
print(series.index("Stranger Things"))  # Imprime o índice do elemento "Game of Thrones" na lista de séries
```

```python
# .count() - Retorna o número de vezes que um elemento específico aparece na lista
print(series.count("Game of Thrones"))  # Imprime o número de vezes que "Game of Thrones" aparece na lista de séries
```
