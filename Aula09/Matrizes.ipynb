# Matriz

```python
# Criando matrix 3x3
matriz = [[1, 2, 3],
          [4, 5, 6],
            [7, 8, 9]]
```

```python
# Criando Matriz apartir de input do usuário
matriz = []
# Loop para preencher a matriz
for i in range(3): #quantidade de colunas
    linha = [] #cria uma nova linha

    for j in range(3): #quantidade de linhas
        valor = int(input(f"Digite o valor para a posição [{i}][{j}]: "))
        linha.append(valor) #adiciona o valor à linha interna
    matriz.append(linha) #adiciona a linha à matriz
# Imprime a matriz
print(matriz)
```

```python
# Criar uma matriz com números aleatórios -> 3x5
import random
matriz = []
for i in range(3): #quantidade de linhas 
    linha = [] #cria uma nova linha

    for j in range(5): #quantidade de colunas
        valor = random.randint(1, 100) #gera um número aleatório entre 1 e 100
        linha.append(valor) #adiciona o valor à linha interna
    matriz.append(linha) #adiciona a linha à matriz

# Imprime a matriz
for linha in matriz:
    print(linha)
```

### Acessando elementos de Matrizes

```python
matriz = [[1, 2, 3],
          [4, 5, 6],
            [7, 8, 9]]
```

```python
print(matriz[0][1]) # Acessa o elemento na primeira linha e segunda coluna (2)
```

```python
print(matriz[1][2])  # Acessa o elemento na primeira linha e terceira coluna (3)
```

### Percorrer linhas de uma Matriz

```python
for linha in matriz:
    print(linha) # Imprime cada linha da matriz
```

```python
for linha in matriz:
    for item in linha:
        print(item) # Imprime cada elemento da matriz
```

```python
# Alterar valores de Matrizes
matriz[0][0] = 20 # Altera o valor na primeira linha e segunda coluna para 20
print(matriz) # Imprime a matriz atualizada
```

```python
# Alterando o valor de uma linha inteira
matriz[2] = [40, 50, 60] # Altera a segunda linha para [40, 50, 60]
print(matriz) # Imprime a matriz atualizada
```

```python
# Alterando coluna de uma matriz
novos_valores = [66, 88, 99] # Novos valores para a terceira coluna

for i in range(len(matriz)):
    matriz[i][2] = novos_valores[i] # Altera o valor da terceira coluna para cada linha

print(matriz) # Imprime a matriz atualizada
```
